```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
pd.set_option('display.height', 1000)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv('../data/modeling_set1.csv')
df = df.dropna(subset=['DaysInHospital'])
df = df.query('age_MISS !=1').drop('age_MISS',axis=1)
df = df.query('sexMISS !=1').drop('sexMISS',axis=1)
```

```python
df.head()
```

```python
tab = pd.crosstab(df['Creditability'], df['Sex & Marital Status'], margins = True)
observed = tab.ix[:-1,:-1] 
stats.chi2_contingency(observed= observed)
```

pval<0.05 so there is a significant relationship between the two variables
