## Packages and settings

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
pd.set_option('display.height', 1000)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv('../data/home_insurance.csv',sep=',')
df = df.query('P1_SEX != "N"') # Remove not given values
df.head()
```

```python
(sns
  .FacetGrid(df, 
             hue='P1_SEX', 
             size=5, 
             xlim=(0, 800))
  .map(sns.kdeplot, 'LAST_ANN_PREM_GROSS', shade=True)
 .add_legend()
)
```

```python
stats.ks_2samp(df.query('P1_SEX == "M"').LAST_ANN_PREM_GROSS, df.query('P1_SEX == "F"').LAST_ANN_PREM_GROSS)
```
